# **Customer Segmentation Analysis**

We'll use K-Means Clustering to segment customers based on their purchase behavior.

```python
# Improt KMeans
from sklearn.cluster import KMeans

# Selecting relevant columns for clustering
features = ['Income', 'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']

# Fill missing values with mean (if necessary)
df[features] = df[features].fillna(df[features].mean())

# Perform K-Means Clustering
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(df[features])

# Visualizing the clusters
sns.scatterplot(data=df, x='Income', y='ID', hue='Cluster', palette='Set1')
plt.title("Customer Segmentation based on Income and ID")
plt.show()
```

Low-income range (under 100,000): Most customers fall into this category. This includes both Cluster 0 (Red) and Cluster 1 (Blue).

Mid-income range (100,000 to 200,000): A few customers are present in this range, mainly from Cluster 2 (Green).

High-income range (200,000 and above): A small number of customers are in this high-income range, all of them belonging to Cluster 2 (Green).

```python
#Income by Recency
# Visualizing the clusters
sns.scatterplot(data=df, x='Income', y='Recency', hue='Cluster', palette='Set2')
plt.title("Customer Segmentation based on Income and Recency")
plt.show()
```

Lower-income groups (Clusters 0 and 1): The vast majority of customers belong to these clusters. They show a wide range of recency values, suggesting a mix of new and more engaged customers.
Higher-income group (Cluster 2): This group is much smaller and consists of higher-income customers with generally lower recency values, suggesting they are either newer or less engaged.
This segmentation shows a clear division between the lower-income majority and a small, high-income customer base. The company could focus on increasing engagement with the higher-income Cluster 2, as they represent potential high-value customers.As well as giving benifits to lower income customers to maintain good customer service and increase sales ratio
